```python
from tkinter import *
import tkinter.filedialog
from tkinter import messagebox


```

```python
#tkinter, abrir janela no python

import os
import pandas as pd
#selecionar o arquivo que a pessoa deve escolher
Janela = Tk()
arquivo = tkinter.filedialog.askopenfilename(title = 'selecione o arquivo com as pesquisas do Youtube para serem descobertos')
Janela.destroy()

caminho =os.getcwd()
buscas_df = pd.read_csv(arquivo, encoding = 'ISO-8859-1', sep=';')
display(buscas_df.head())
```

```python
import time

for coluna in buscas_df:
    print(coluna)
    print(len(buscas_df))
    for linha in range(len(buscas_df)):    
        print(buscas_df[coluna][linha])
        pesquisa=buscas_df[coluna][linha]
        #coletar já a pesquisa que eu quero da lista
        navegador.get(f"https://www.youtube.com/results?search_query={pesquisa}")
    
        #o id neste caso, depois de inspecionar no chrome, é, video-title
        
        time.sleep(10)


        lista_videos = navegador.find_elements(By.ID, 'video-title')

        print('lista de vídeos aula',len(lista_videos))

        #quero pegar o atributo vídeo
        #quero carregar mais vídeos e forçar para carregar na página os links
        
        #vou dar scroll na tela
        #aba do console, na parte do inspecionar
        #window.scroll(0,100)
        #window.scroll(0,1000)
        #window.scroll(0,10000)
        #vai descendo
        
        #comando para rodar o javascript
        #podemos rodar dentro do chrome, no inspecionar, console
        
        navegador.execute_script('window.scroll(0,1000)')


        

        for i in range(30):
        #dar 1000 a 1000
            qtde_scroll= i * 1000 
            time.sleep(5)
            navegador.execute_script(f'window.scroll(0,{qtde_scroll})')



        hrefs=[]
        for video in lista_videos:
            meu_link=video.get_attribute("href")
            print(meu_link)
            hrefs.append(meu_link)
            #pegar dados pelo get_attribute
        print(len(lista_videos))

        #salvando o resultado em um csv

        hrefs_df = pd.DataFrame(hrefs)
        hrefs_df.to_csv(rf'CanaisProntos{pesquisa}.csv', sep=',', encoding='utf-8')
        display(hrefs_df)


#avisando ao cliente que o resultado foi feito
janela = Tk()

messagebox.showinfo('Status do programa de pesquisa do you tube', 'Programa concluído com sucesso. Pegue o arquivo em .csv, para ser visualizado pelo excel na pasta do programa' )
janela.destroy()
```
